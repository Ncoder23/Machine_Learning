## Plotting Histogram using Matplotlib Libray:

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data['Pregnancies'])

plt.xlabel('Pregnancies')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
plt.plot(data['Pregnancies'])

plt.xlabel('Pregnancies')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
plt.hist(data['Glucose'])

plt.xlabel('Glucose')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
plt.plot(data['Glucose'])

plt.xlabel('Glucose')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
plt.hist(data['BloodPressure'])
plt.xlabel('bloodPressure')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
plt.plot(data['BloodPressure'])

plt.xlabel('BloodPressure')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
plt.hist(data['SkinThickness'])

plt.xlabel('SkinThickness')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
plt.plot(data['SkinThickness'])

plt.xlabel('SkinThickness')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
plt.hist(data['Insulin'], bins=5)

plt.xlabel('Insulin')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
plt.plot(data['Insulin'])

plt.xlabel('Insulin')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
plt.hist(data['BMI'], bins=5)

plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
plt.plot(data['BMI'])

plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
plt.hist(data['DiabetesPedigreeFunction'], bins=5)

plt.xlabel('DiabetesPeigreeFunction')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
plt.plot(data['DiabetesPedigreeFunction'])

plt.xlabel('DiabetesPedigreeFunction')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
plt.hist(data['Age'], bins=15)

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram')
```

```python
plt.plot(data['Age'])

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram')
```
