# Practical 10 : Implement Convolution Neural network on FER-2013 dataset of Kaggle using Python

```python
import tensorflow
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model,Sequential
from tensorflow.keras.layers import Input,Conv2D,Dropout,Dense,MaxPooling2D,Flatten
```

```python
labels = ['angry','disgust','fear','happy','neutral','sad','surprise']
```

```python
data_gen = ImageDataGenerator(rescale=1/255)
data = data_gen.flow_from_directory('archive/test/',target_size=(128,128))
```

```python
data.class_indices
```

```python
data.image_shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(data[0][0][0])
```

```python
model = Sequential([
    Input(shape=(128,128,3)),
    Conv2D(32,3,activation='relu'),
    Conv2D(32,3,activation='relu'),
    MaxPooling2D(3),
    Conv2D(64,3,activation='relu'),
    Conv2D(64,3,activation='relu'),
    MaxPooling2D(3),
    Flatten(),
    Dense(128,activation='relu'),
    Dense(64,activation='relu'),
    Dense(7,activation='softmax')
])
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics='accuracy')
model.summary()
```

```python
model.fit(data,epochs=10)
```
