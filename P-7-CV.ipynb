```python
from skimage.feature import hog
```

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
ori = cv2.imread('/content/drive/MyDrive/Colab Notebooks/N.jpg')
img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/N.jpg")
```

```python
 
_, hog_image = hog(img, orientations=8, pixels_per_cell=(16, 16), cells_per_block=(1, 1), visualize=True, multichannel=True)
cv2_imshow(ori)
cv2_imshow(hog_image)
```

```python
import cv2 as cv
gray_scale = cv.cvtColor(ori, cv.COLOR_BGR2GRAY)

#initialize SIFT object
sift = cv.xfeatures2d.SIFT_create()

#detect keypoints
keypoints, _= sift.detectAndCompute(ori, None)

#draw keypoints
sift_image = cv.drawKeypoints(gray_scale, keypoints, None)

cv2_imshow(sift_image)
cv.waitKey(0)
```
